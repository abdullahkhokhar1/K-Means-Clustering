# K-Means Clustering on Mall Customers Dataset

```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#Loading Dataset
df = pd.read_csv('Mall_Customers.csv')
```

```python
#Displaying the first few rows of the dataset
df.head()
```

```python
#Extracting relevant features for clustering
features = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
#Standardizing the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
#Function to perform K-Means and return cluster centers and value counts
def perform_kmeans(n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(features_scaled)
    cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
    value_counts = pd.Series(kmeans.labels_).value_counts().sort_index()
    return cluster_centers, value_counts
```

```python
#Performing K-Means for 3, 4, and 5 clusters
cluster_centers_3, value_counts_3 = perform_kmeans(3)
cluster_centers_4, value_counts_4 = perform_kmeans(4)
cluster_centers_5, value_counts_5 = perform_kmeans(5)
```

```python
#Calculating the statistical values
min_age = df['Age'].min()
max_age = df['Age'].max()
avg_age = df['Age'].mean()
min_income = df['Annual Income (k$)'].min()
max_income = df['Annual Income (k$)'].max()
avg_income = df['Annual Income (k$)'].mean()
min_score = df['Spending Score (1-100)'].min()
max_score = df['Spending Score (1-100)'].max()
avg_score = df['Spending Score (1-100)'].mean()
unique_ages = df['Age'].nunique()
```

```python
#Printing the results
print("Minimum Age:", min_age)
print("Maximum Age:", max_age)
print("Average Age:", avg_age)
print("Minimum Annual Income:", min_income)
print("Maximum Annual Income:", max_income)
print("Average Annual Income:", avg_income)
print("Minimum Spending Score:", min_score)
print("Maximum Spending Score:", max_score)
print("Average Spending Score:", avg_score)
print("Number of unique ages:", unique_ages)
print("\nCluster Centers with 3 Clusters:\n", cluster_centers_3)
print("Value Counts with 3 Clusters:\n", value_counts_3)
print("\nCluster Centers with 4 Clusters:\n", cluster_centers_4)
print("Value Counts with 4 Clusters:\n", value_counts_4)
print("\nCluster Centers with 5 Clusters:\n", cluster_centers_5)
print("Value Counts with 5 Clusters:\n", value_counts_5)
```

```python
#Plots the clusters for visualization
def plot_clusters(n_clusters, features, labels, title):
    plt.figure(figsize=(10, 6))
    plt.scatter(features[:, 0], features[:, 1], c=labels, cmap='viridis')
    plt.title(title)
    plt.xlabel('Age')
    plt.ylabel('Annual Income (k$)')
    plt.show()
```

```python
#Performing K-Means again to get the labels for plotting
kmeans_3 = KMeans(n_clusters=3, random_state=42).fit(features_scaled)
kmeans_4 = KMeans(n_clusters=4, random_state=42).fit(features_scaled)
kmeans_5 = KMeans(n_clusters=5, random_state=42).fit(features_scaled)
```

```python
#Plotting the graphs
plot_clusters(3, features_scaled, kmeans_3.labels_, '3 Clusters Graph')
plot_clusters(4, features_scaled, kmeans_4.labels_, '4 Clusters Graph')
plot_clusters(5, features_scaled, kmeans_5.labels_, '5 Clusters Graph')
```
